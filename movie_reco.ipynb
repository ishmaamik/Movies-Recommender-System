```python
#Genres here is basically a string of list of dictionaries so first we need to convert it to List
#so we convert the string to a list using the ast and ast's literal_eval
#append basically adds to end of list

import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!= 3:
         L.append(i['name'])
         counter+=1
        else:
            break
    return L
    
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#indentation of break is important, if break used exactly below if it checks if condition, not matches so break with first element.
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
         L.append(i['name'])
         break
    return L
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#lambda is anonymous function for just a line call and split basically separates into words and individual lists per row.
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#gives no space to those with space in between
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview']+ movies['genres'] +movies['keywords']+ movies['cast']+ movies['crew']
```

```python
#converts list of strings into strings themselves per row
new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
```

```python
#choose from ishuenv kernel in top right corner
#choose 5000 most common words so max_features=5000
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000, stop_words='english')
```

```python
#scipy sparse matrix by default, so convert it to array 
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
#apply stemming to make similar words all the same e.g. loving, lovely, lovable to 'love' so all those are treated as the same common word
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python


def stem(text):
    y= []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#4806 by 4806 since each movie with every movie- 
similarity= cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#only the values but without indices loses the positions so we use enumerate with sort
#similarity[0] means the movie, x[1] means with respect to the second column sorting should be done which is cosine similarities
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
#new_df['title']==movie finds from new_df the object of the movie and then this object put into new_df[movie] and from here the index[0] or 
#movie_id is extracted
def recommend(movie):
  movie_index= new_df[new_df['title']== movie].index[0]
  distances= similarity[movie_index] #finds the similarity cosine list of the given movie
  movies_list= sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]  #distances = similarity[movie_index] so it sorts the similarity list of that movie

  for i in movies_list:
      print(new_df.iloc[i[0]].title)   #i[0] gives the first column of movies_list which is the movie ids
#iloc gives the exact position
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```
